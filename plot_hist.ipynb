```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
bpd_file_paths = glob('outs/cifar10_subvp_likelihood_tw/eval_cifar/test_ckpt_26_bpd_*.npz') 
bpd_file_paths = sorted(bpd_file_paths)
bpds = []
for p in bpd_file_paths:
    tmp = np.load(p)
    bpds.append(tmp['arr_0'])
bpds = np.concatenate(bpds)
cifar_nll = bpds
```

```python
bpd_file_paths = glob('outs/cifar10_subvp_likelihood_tw/test_eval_svhn/test_ckpt_26_bpd_*.npz') 
bpd_file_paths = sorted(bpd_file_paths)
bpds = []
for p in bpd_file_paths:
    tmp = np.load(p)
    bpds.append(tmp['arr_0'])
bpds = np.concatenate(bpds)
svhn_nll = bpds
```

```python
bpd_file_paths = glob('outs/cifar10_subvp_likelihood_tw/test_eval_svhn_shuffle_8x8/test_ckpt_26_bpd_*.npz') 
bpd_file_paths = sorted(bpd_file_paths)
bpds = []
for p in bpd_file_paths:
    tmp = np.load(p)
    bpds.append(tmp['arr_0'])
bpds = np.concatenate(bpds)
shuffle_svhn_nll = bpds
```

```python
plt.figure(figsize=(20,10))
plt.title("Histogram Score flow - trained on CIFAR10")
plt.xlabel("Negative bits per dimension")
plt.hist(-svhn_nll, label="SVHN", density=True, bins=30)
plt.hist(-cifar_nll, label="CIFAR10", density=True, bins=30)
plt.hist(-shuffle_svhn_nll, label="Shuffle 8x8 SVHN", density=True, bins=30)
plt.legend()
plt.show()
# plt.savefig("images/histogram_glow_cifar_svhn.png", dpi=300)
```
